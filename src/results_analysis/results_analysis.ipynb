## Plot graph for overall accuracies

#### Normal Cross Entropy Loss

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#choose accuracy to plot
plot_df = acc_nl
```

```python
fig, axes = plt.subplots(
    nrows=1, 
    ncols=3, 
    sharey=True,
    figsize=(18,7)
)
ax1, ax2, ax3 = axes
color = {
    'FEVER': '#9abbff',
    'PUBHEALTH': 'gainsboro',
    'CLIMATE-FEVER': '#f7a889'
}

def make_plot(_df, ax, title=None, **kwargs):
    sns.barplot(
        ax=ax,
        data = _df,
        x="model",
        y="correctly_classified",
        order = ['BERT', 'RoBERTa', 'ALBERT', 'SciBERT', 'BioBERT'],
        hue="test_dataset", errorbar=None,
        palette=color,
        **kwargs
    )

    ax.set_xlabel(None)
    ax.set_ylabel(None)
    ax.set_ylim(0, 99)
    ax.grid(axis='y', alpha=0.5)
    ax.legend_.remove()
    ax.set_axisbelow(True)
    ax.set_title(title, fontsize=20, pad=14)
    ax.tick_params(axis='x', labelsize=15, pad=10)

    return ax


#plot first graph
_df = plot_df[plot_df['train_dataset'] == 'Fine-tuned on FEVER']
ax1 = make_plot(_df, ax1, 
        title="Fine-tuned on FEVER", 
        hue_order = ['FEVER', 'PUBHEALTH', 'CLIMATE-FEVER'])
ax1.set_ylabel('Accuracy (%)', fontsize=20, labelpad=14)


#plot second graph
_df = plot_df[plot_df['train_dataset'] == 'Fine-tuned on PUBHEALTH']
ax2 = make_plot(_df, ax2, 
        title="Fine-tuned on PUBHEALTH", 
        hue_order = ['PUBHEALTH', 'FEVER', 'CLIMATE-FEVER'])

#plot 3rd graph
_df = plot_df[plot_df['train_dataset'] == 'Fine-tuned on CLIMATE-FEVER']
ax3 = make_plot(_df, ax3, 
        title="Fine-tuned on CLIMATE-FEVER", 
        hue_order = ['CLIMATE-FEVER', 'FEVER', 'PUBHEALTH'])

ax2.legend(
    frameon=False, 
    loc="lower center", 
    ncol=3,
    fontsize=20,
    mode="expand",
    bbox_to_anchor=(-0.2,-0.25, 1.4, 1),
    handletextpad=0.5
)

#Show bar values
for axis in fig.axes:
    for i in axis.containers:
        axis.bar_label(i, fmt='{:,.0f}', fontsize=12)


fig.tight_layout()
```

```python
fig.savefig("overall_accuracy_normal.png")
```

#### Weighted Cross Entropy Loss

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
fig, axes = plt.subplots(
    nrows=1, 
    ncols=3, 
    sharey=True,
    figsize=(18,7)
)
ax1, ax2, ax3 = axes
color = {
    'FEVER': '#9abbff',
    'PUBHEALTH': 'gainsboro',
    'CLIMATE-FEVER': '#f7a889'
}

def make_plot(_df, ax, title=None, **kwargs):
    """Plot v2 results
    """
    sns.barplot(
        ax=ax,
        data = _df,
        x="model",
        y="correctly_classified",
        order = ['BERT', 'RoBERTa', 'ALBERT', 'SciBERT', 'BioBERT'],
        hue="test_dataset", errorbar=None,
        palette=color,
        # linestyle="-",
        # edgecolor="grey",
        **kwargs
    )

    ax.set_xlabel(None)
    ax.set_ylabel(None)
    ax.set_ylim(0, 99)
    ax.grid(axis='y', alpha=0.5)
    ax.legend_.remove()
    ax.set_axisbelow(True)
    ax.set_title(title, fontsize=20, pad=14)
    ax.tick_params(axis='x', labelsize=15, pad=10)

    return ax

def overlay_plot(_df, ax, title=None, **kwargs):
    """Plot v1 results
    """
    sns.barplot(
        ax=ax,
        data = _df,
        x="model",
        y="correctly_classified",
        order = ['BERT', 'RoBERTa', 'ALBERT', 'SciBERT', 'BioBERT'],
        hue="test_dataset", errorbar=None,
        palette=color,
        facecolor=(0,0,0,0),
        edgecolor="black",
        # linestyle="--",
        linestyle=(0, (5, 5)),
        # linestyle=(0, (3, 5, 1, 5)),
        # hatch="/",
        **kwargs
    )
    ax.set_xlabel(None)
    ax.set_ylabel(None)
    ax.legend_.remove()
    return ax

#plot first graph
_df = acc_wl[acc_wl['train_dataset'] == 'Fine-tuned on FEVER']
ax1 = make_plot(_df, ax1, 
        title="Fine-tuned on FEVER", 
        hue_order = ['FEVER', 'PUBHEALTH', 'CLIMATE-FEVER'])

_df = acc_nl[acc_nl['train_dataset'] == 'Fine-tuned on FEVER']
overlay_plot(_df, ax1, hue_order = ['FEVER', 'PUBHEALTH', 'CLIMATE-FEVER'])

ax1.set_ylabel('Accuracy (%)', fontsize=20, labelpad=14)


#plot second graph
_df = acc_wl[acc_wl['train_dataset'] == 'Fine-tuned on PUBHEALTH']
ax2 = make_plot(_df, ax2, 
        title="Fine-tuned on PUBHEALTH", 
        hue_order = ['PUBHEALTH', 'FEVER', 'CLIMATE-FEVER'])

_df = acc_nl[acc_nl['train_dataset'] == 'Fine-tuned on PUBHEALTH']
overlay_plot(_df, ax2, hue_order = ['PUBHEALTH', 'FEVER', 'CLIMATE-FEVER'])

#plot 3rd graph
_df = acc_wl[acc_wl['train_dataset'] == 'Fine-tuned on CLIMATE-FEVER']
ax3 = make_plot(_df, ax3, 
        title="Fine-tuned on CLIMATE-FEVER", 
        hue_order = ['CLIMATE-FEVER', 'FEVER', 'PUBHEALTH'])

_df = acc_nl[acc_nl['train_dataset'] == 'Fine-tuned on CLIMATE-FEVER']
overlay_plot(_df, ax3, hue_order = ['CLIMATE-FEVER', 'FEVER', 'PUBHEALTH'])


# make legend
handles, labels = ax2.get_legend_handles_labels()
fig.legend(
    handles[0:3],
    labels[0:3],
    frameon=False, 
    loc="upper left", 
    ncol=1,
    fontsize=20,
    # mode="expand",
    handletextpad=0.5,
    bbox_to_anchor=(0.25, -0.52, 0.5, 0.5)
)



normal = mpatches.Patch(fill=False, label="Cross Entropy Loss", linestyle=(0, (5,5)))
weighted = mpatches.Patch(fill=False, label="Weighted Cross Entropy Loss", linestyle="-")

fig.legend(
    handles=[normal, weighted],
    loc="upper right",
    frameon=False,
    ncol=1,
    fontsize=20,
    handletextpad=1,
    bbox_to_anchor=(0.3, -0.52, 0.5, 0.5)
)

#Show bar values
# for axis in fig.axes:
#     for i in axis.containers[0:3]:
#         axis.bar_label(i, fmt='{:,.0f}', fontsize=12)


fig.tight_layout()
```

## Plot graph for label analysis

```python
import seaborn as sns
```

```python
def plot_label_graph(stats):
    _df = stats.copy()

    order = ["BERT", "RoBERTa", "ALBERT", "SciBERT", "BioBERT"]
    _df['model'] = pd.Categorical(_df.model, categories=order, ordered=True)
    _df.sort_values('model')

    _df['train_dataset'].replace({
        'Fine-tuned on FEVER': 'Fine-tuned \n on FEVER',
        'Fine-tuned on PUBHEALTH': 'Fine-tuned on \n PUBHEALTH',
        'Fine-tuned on CLIMATE-FEVER': 'Fine-tuned on \n CLIMATE-FEVER'
    }, inplace=True)
    
    sns.set_theme(style="ticks", font_scale=1.2)
    sns.set_style("ticks", { 'axes.grid': False })

    fontsize=18

    g = sns.FacetGrid(
        _df, 
        row='train_dataset',
        col='test_dataset',
        height=4, aspect=1.6, 
        despine=False,
        margin_titles=True,
        row_order = ['Fine-tuned \n on FEVER', 'Fine-tuned on \n PUBHEALTH', 'Fine-tuned on \n CLIMATE-FEVER'],
        col_order = ['FEVER', 'PUBHEALTH', 'CLIMATE-FEVER']
    )

    g.map_dataframe(
        sns.lineplot, 
        data=_df, 
        x='model',
        y='accuracy', 
        hue='labels',
        style='labels',
        markers=True,
        err_style=None,
        palette="muted",
        markersize=10,  #default 6
    )

    g.add_legend(loc="lower center", ncol=3, fontsize=fontsize, frameon=True)
    g.set_titles(
        col_template="{col_name}", 
        row_template="{row_name}", 
        size=fontsize, 
        horizontalalignment="center"
    )
    g.set_xlabels("")
    g.set_ylabels("Accuracy (%)", fontsize=fontsize)
    g.set_xticklabels(
        labels=["BERT", "RoBERTa", "ALBERT", "SciBERT", "BioBERT"], 
        rotation=20
    )

    g.figure.subplots_adjust(
        wspace=.05, 
        hspace=.2,
        bottom=0.12
    )

    subplot_bg = "lightyellow"
    g.axes[0][0].set_facecolor(subplot_bg)
    g.axes[1][1].set_facecolor(subplot_bg)
    g.axes[2][2].set_facecolor(subplot_bg)
```

```python
plot_label_graph(stats_nl)
```

```python
plot_label_graph(stats_wl)
```
